● WWW – What content resonates with her? What are the themes and frames that drive engagement?

```python
# Which articles are the most popular per age group?

# excluding the homepage hits
audience = www_formed[~www_formed.page_title.str.contains("Celebrity Style and Fashion Trend Coverage")]

# summing the Page views
audience_reads = audience.groupby(['Age', 'page_title']).sum()

# droping the date column, as it was also being summed
audience_reading = audience_reads.drop(['Date'], axis=1)

# final result
audience_reading.sort_values(by=['Pageviews'], ascending=False).head(20)
```

```python
# another way to do the same thing as above
audience_reads.pivot_table(audience_reads,index=['Age', 'page_title'])\
.sort_values(by=['Pageviews'], ascending=False).head(5)
```

```python
# Total page views, per article, by each age group through each day of the month
audience.groupby(['Date', 'Age', 'page_title']).sum()
```
